```python
from datetime import datetime, timedelta
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import random
import psycopg2 as pg2
import pandas as pd
# Always make it pretty.
plt.style.use('ggplot')
```

```python
def bootstrap(start_date, end_date, iterations, column, wrap = False):
    conn = pg2.connect(host = 'localhost', user='postgres', password='Password', dbname = "weatherdata")
    cur = conn.cursor()
    retVal = pd.DataFrame(columns=[column])
    for i in range(iterations):
        val = pd.DataFrame(columns=[column])
        while val.empty == True:
            year = random.randint(start_date.year, end_date.year)
            if wrap: #Ensuring that if I put in values like Dec to Jan
                flip = random.randint(0,2)
                if flip ==0:
                    month = random.randint(start_date.month, 12) 
                else:
                    month = random.randint(1, end_date.month)
            elif start_date.year - end_date.year >1: #If they're more than a year apar
                month = random.randint(1,12)
            else:
                month = random.randint(start_date.month, end_date.month)
            day = random.randint(1, 31) #Selecting a day to pull
            #cur.execute("SELECT %s FROM weather WHERE year = %s AND month = %s AND day = %s", (column, year, month, day))
            val = pd.read_sql_query(("SELECT %s FROM weather WHERE year = %s AND month = %s AND day = %s"%(column, year, month, day)), conn)
            #print(val)
        retVal = retVal.append(val)
    return retVal
```

```python
current_date = datetime(year=2019, month = 6, day =1)
end_date = datetime(year=2019, month = 8, day = 31)
print(bootstrap(current_date, end_date, 100, 'actualdailyavgtemp', False))
```

```python
def plotSamples(setA, setB):
    meanA = stats.mean(setA)
    total = 0
    for el in setA:
        total+=((el-meanA)**2) #Calculating sample distribution values for A
    stdA = ((total/len(setA))**(1/2))

    meanB = stats.mean(setB)
    total = 0
    for el in setB: #Calculating sample values for B
        total+=((el-meanB)**2)
    stdB = ((total/len(setB))**(1/2))
    
    distA = stats.norm(meanA, stdA)
    distB = stats.norm(meanB, stdB)
    lb = stats.min([stats.min(setA), stats.min(setB)]) #Choosing what values I should use as my min and max x values
    ub = stats.max([stats.max(setA), stats.max(setB)])
    x = np.linspace(lb, ub, 500)
    
    fig, ax = plt.subplots() #Plotting my distributions
    ax.plot(x, distA.pdf(x), label = "DistA")
    ax.plot(x, distB.pdf(x), label = "DistB") 
```

```python
def plotDistrib(distA, testval):
    fig, ax = plt.subplots(figsize = (8,6))
    x = np.linspace((distA.mean() - (4*distA.std())), (distA.mean() + (4*distA.std())), 200)
    ax.plot(x, distA.pdf(x))
    ax.vlines(distA.ppf(.95), 0, 1, label = '95% critical value')
    ax.vlines(testval, 0, 1, color = 'red', label = 'last 4 years average')
    ax.legend()
```

```python
def plot2Distrib(distA, distB):
    fig,ax =plt.subplots(figsize = (8,6))
    x = np.linspace((min((distA.mean() - (4*distA.std())), (distB.mean()-(4*distB.std())))), 
                    (max((distA.mean() + (4*distA.std())), (distB.mean()+(4*distB.std())))), 200)
    ax.plot(x, distA.pdf(x), label ="1970s rainfaill")
    ax.plot(x, distB.pdf(x), label ="2010s rainfaill")
    ax.legend()
```

```python
EHD = pd.read_sql_query('SELECT actualhightemp - histavghightemp AS diff FROM weather WHERE year IN (1970, 1971, 1972, 1973);', conn)
LHD = pd.read_sql_query('SELECT actualhightemp - histavghightemp AS diff FROM weather WHERE year IN (2016, 2017, 2018, 2019, 2020);', conn)
EHDist = stats.norm(EHD.mean(), EHD.sem())
LHDist = stats.norm(LHD.mean(), LHD.sem())
plot2Distrib(EHDist, LHDist)
```

```python
def plotfourDist(distA, distB, distC, distD, valA, valB, valC, valD):
    fig, ax = plt.subplots(2,2, figsize = (12,8))
    Dlist = [distA, distB, distC, distD]
    Vlist = [valA, valB, valC, valD]
    for axis, dist, val in zip(ax.flatten(), Dlist, Vlist):
        x = np.linspace((dist.mean() - (4*dist.std())), (dist.mean() + (4*dist.std())), 200)
        axis.plot(x, dist.pdf(x))
        axis.vlines(dist.ppf(.95), 0, 1, label ='95% critical value')
        axis.vlines(val, 0, 1, color='r', label = 'last 4 years average')
        axis.legend()
    ax[0][0].set_title('Summer')
    ax[0][1].set_title('Fall')
    ax[1][0].set_title('Winter')
    ax[1][1].set_title('Spring')
        
```
